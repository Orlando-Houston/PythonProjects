```python
house_data = pd.read_csv('C:/Users/aozde/OneDrive/Documents/CapstoneProject/HouseData/Home/house_data.csv')
h_d=house_data.copy()
pd.set_option("display.float_format", "{:.2f}".format)
h_d.head()
```

```python
#h_d =h_d.iloc[:,1:len(h_d)]
#h_d.head()
```

```python
h_d=h_d.drop('date',axis=1)
h_d.head()
```

```python
#number of bathrooms is a float value. convert it to integer
h_d['bathrooms']=h_d['bathrooms'].astype(int)
h_d.head()
```

```python
# count
len(h_d.zipcode.unique())
```

```python
from sklearn.preprocessing import LabelEncoder

lb_make = LabelEncoder()
h_d["zipcode"] = lb_make.fit_transform(h_d["zipcode"])
h_d.zipcode.dtypes
```

```python
dummies =pd.get_dummies(h_d.zipcode)
dummies.head()
```

```python
D_h_d=pd.concat([h_d,dummies],axis='columns')
D_h_d.head(1)
```

```python
D_h_d_ = D_h_d.drop('zipcode',axis ='columns')
D_h_d_.head(3)
```

```python
cols = ['price','bedrooms','bathrooms','sqft_living','sqft_lot','floors',60,61,62,63,64,65,66,67,68,69]
h_d_c = D_h_d_[cols]
h_d_c.head()
```

```python
h_d_c.info
```

# Create Modeling
## Supervised Learning
### train-test split

```python
## independes variables
dataset_features = h_d_c.drop(['price'], axis=1)
## target/depends variable
dataset_labels = h_d_c['price']
```

```python
dataset_features
```

```python
dataset_labels
```

```python
from sklearn.model_selection import train_test_split

train_features,test_features,train_labels,test_labels= train_test_split(dataset_features,dataset_labels,test_size=0.2, random_state=21)
```

## Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

regressorR_f = RandomForestRegressor(n_estimators=200, random_state=0)

regressorR_f.fit(train_features, train_labels)
```

```python
regressorR_f.score(dataset_features,dataset_labels)
```

```python
predicted_price = regressorR_f.predict(test_features)
predicted_price
```

```python
from sklearn import metrics
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error( test_labels, predicted_price)) 
print('Mean Squared Error:', metrics.mean_squared_error(test_labels,predicted_price))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(test_labels, predicted_price)))
```

```python
def predict_price(location,sqft_living,bedrooms,bathrooms):
    loc_index = np.where(X.columns ==location)[0][0]
    
    x = np.zeros(len(X.columns))
    x[0] =sqft_living
    x[1]=bedrooms 
    x[2]=bathrooms
    
    if loc_index >= 0:
        x[loc_index] =1
        
    return regressorR_f.predict([x])[0]
```

```python
predict_price(67,2500,3,2)
```
