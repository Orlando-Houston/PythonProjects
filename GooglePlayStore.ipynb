* df’ deki eksik verile
* bfillna() ve dropna() metotları
* dropna() metodu NaN değerler bulunduran satırları veya sütunları silmemizi sağlıyor

```python
df["Rating"].fillna(df.groupby("Category") ["Rating"]. transform("mean"), inplace=True)
                               
print(df.isnull().sum().sort_values(ascending=False))
```

```python
#11 tane Nan deger var  /dropna metodu ile silmek
df=df.dropna()
print(df.isnull().sum().sort_values(ascending=False))
```
